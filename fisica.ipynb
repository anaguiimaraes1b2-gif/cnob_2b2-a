 ## função do deslocamento

```
def calculaDesloc(inicialS, deltaT, v):
  d = inicialS + (v*deltaT)
  return d
```

```
print("Deslocamento= ",calculaDesloc(0, 15, 30))
```

#Fórmulas do Movimento Uniformemente Variado (MUV)

```
def calvelcidadehoraria(vInicil, acel, tempo):
  vh = vInicial + (acel * tempo)
  return vh
```

```
print("velocidade horária: ", calvelcidadehoraria(12, 25,78))
```
